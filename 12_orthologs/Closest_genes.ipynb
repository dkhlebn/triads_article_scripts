```python
with open("orthological_rd.tsv", 'r') as inp:
    s = set()
    for line in inp:
        line = line.strip().split('\t')
        orth1 = '\t'.join([line[0], line[1], line[2], line[3], line[10]])
        orth2 = '\t'.join([line[0], line[5], line[6], line[7], line[10]])
        if orth1 in s or orth2 in s:
            continue
        else:
            line = [line[0], line[3], line[7], line[9], line[10]]
            print(' & '.join(line), end=" \\\\ \hline\n")
            s.add(orth1)
            s.add(orth2)
```
